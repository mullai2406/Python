```python
import pandas as pd
import numpy as np
```

```python
#Assign the Data(Excel File) to the Dataset
dataset = pd.read_csv("salary.csv")
```

```python
dataset
```

```python
# Assign the dataset data to the "ind" Variable
ind = dataset[['yrs']]
ind
# Assign the dataset data to the "dep" Variable
dep = dataset [['salary']]
dep
```

```python
#Importing the Train test slpit function from the sklearn library
from sklearn.model_selection import train_test_split
```

```python
#Assigning the ind and dep data to the (training test function variables) and the test data to 30%
x_train,x_test,y_train,y_test = train_test_split(ind,dep,test_size=0.30,random_state=0)

#importing the Liner REgression function from the sklearn
from sklearn.linear_model import LinearRegression

#Assigning the Liner Regression Class to the variable as "reg"
reg=LinearRegression()

# Fitting Function,  Fitting training set input and training set output data to the regression 
reg.fit(x_train, y_train)
```

```python
# Now the regression coefficient is assigned to the weight variable
weight=reg.coef_
```

```python
weight
```

```python
# Now the regression intercept is assigned to the weight variable
bais=reg.intercept_
```

```python
bais
```

```python
# Assigning the Predit Value to the Variable (Cross Check Varaible)  
y_pred=reg.predict(x_test)
```

```python
#importing the Librabry r2_Score from sklearn
from sklearn.metrics import r2_score

#calcuating the r_score value from y_test and y_predit
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# Importing the Function "Pickle" to save our model in file
import pickle

# Assigning the File name to save our model vida variable "filename"
filename="fin_model_liner.sav"
```

```python
# Saving our model in file in Binary format "Write binary"
pickle.dump(reg,open(filename,'wb'))
```

```python
# Loading our Model from File "Read Binary"
loaded_model=pickle.load(open("fin_model_liner.sav",'rb'))

# Testing the model in realtime
result=loaded_model.predict([[28]])
```
